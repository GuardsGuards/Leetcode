LeetCode 448: Find All Numbers Disappeared in an Array with the in-place marking trick: use each value v to index v-1 and flip that spot negative; positives at the end are the missing numbers.

---

## The Problem

You have an array of length `n` containing integers from `1` to `n`. Some are missing, some may repeat. You want to find which ones are missing **without extra memory**.

---

## The Core Idea

Every number `v` in `1..n` has a **home index** at `v-1`.
Example:

* `1` belongs at index `0`
* `2` belongs at index `1`
* …
* `n` belongs at index `n-1`

So if you walk through the array and, for each number `v`, mark `nums[v-1]` somehow, then at the end you know which positions were “touched” by numbers in the array.

---

## The Marker

Integers are already signed (positive/negative).

* Initially all are positive.
* If you see a number `v`, you **flip `nums[v-1]` to negative**.
* Negative = “this number has been seen at least once.”
* If it’s already negative, you leave it (duplicate encountered).

At the end:

* Any index `i` that is still positive means its number `(i+1)` was never seen.

---

## Example Walkthrough

Input:

```
nums = [4,3,2,7,8,2,3,1]
```

Step 1: Look at `4` → home index `3` → flip `nums[3]` → array becomes `[4,3,2,-7,8,2,3,1]`.
Step 2: Look at `3` → index `2` → flip → `[4,3,-2,-7,8,2,3,1]`.
Step 3: Look at `2` → index `1` → flip → `[4,-3,-2,-7,8,2,3,1]`.
Step 4: Look at `7` → index `6` → flip → `[4,-3,-2,-7,8,2,-3,1]`.
Step 5: Look at `8` → index `7` → flip → `[4,-3,-2,-7,8,2,-3,-1]`.
Step 6: Look at `2` again → index `1` → already negative, skip.
Step 7: Look at `3` again → index `2` → already negative, skip.
Step 8: Look at `1` → index `0` → flip → `[-4,-3,-2,-7,8,2,-3,-1]`.

Now scan:

* Indices `4` and `5` are positive → numbers `5` and `6` are missing.

Output = `[5,6]`.

---

## Why It Works

1. **Guaranteed mapping**: Each number is between `1` and `n`, so its index `v-1` is valid.
2. **Stable marker**: Negating preserves the number (you can still recover it with `abs`) while giving you a one-bit “seen” flag.
3. **In-place**: No extra set/dictionary needed. Memory stays O(1).
4. **Duplicates handled**: Multiple hits just leave the target negative.

---

### **USPS DS/DE (fraud + ops data)**

Packages are scanned at certain checkpoints (say 1–10). You expect each package to hit all checkpoints in sequence. Reality: scanners misfire or data is dropped.
 Apply the exact same trick: feed in the observed checkpoint IDs for a package, flip the seen ones, and the function returns **which checkpoints never got scanned**. That becomes your operational alert: “package 782 never hit checkpoint 5 or 7.”

---

### **Most Relevant Elsewhere: Inventory / Attendance**

Any situation where you expect **a fixed roster 1..N but only observe some subset**:

* Retail: 1–100 items should be on the shelf, scanners report 92, missing 8 are flagged.
* Classroom: 1–30 students, attendance IDs recorded, missing IDs tell you absentees.
* Cloud jobs: 1–500 batch IDs expected, only 485 finished logs present, missing 15 = failures.

 Same exact problem: expected full range, return the absent IDs.

---

The core utility is: **detect gaps in a supposed complete set without using extra memory** — just flip markers in-place.




````python
from typing import List

def find(nums: List[int]) -> List[int]:

    # Mark seen indices as negative
    for i in range(len(nums)):
        temp = abs(nums[i]) - 1
        if nums[temp] > 0:
            nums[temp] *= -1

    # Collect indices tht stayed positive => missing values
    res = []
    for i,n in enumerate(nums):
        if n > 0:
            res.append(i + 1)

    return res

# Example
nums = [4, 3, 2, 7, 8, 2, 3, 1]
print(find(nums))
````
